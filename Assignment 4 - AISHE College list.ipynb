# Problem statement

Go to URL: http://aishe.gov.in/aishe/institutionAisheCode?universityId=-1&flag=true&districtCode=-1&universitytypeId=-1&stateId=29&instituteTypeId=-1&institutionType=college&d-4030398-p=1

Give the parameters and collect all the colleges data by rotating over all the pages for the state Karnataka and store it into a data frame using python in ipynb format.



# Importing the libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from random import randint
from time import sleep
```

### Using bs4 to navigate the URL and scrape table head

```python
for page in range(1,2): 
    req = requests.get("http://aishe.gov.in/aishe/institutionAisheCode?universityId=-1&flag=true&districtCode=-1&universitytypeId=-1&stateId=29&instituteTypeId=-1&institutionType=college&d-4030398-p={}".format(page),verify=False)
    soup = bs(req.text, 'html.parser')
    table = soup.find("table", attrs={"class": "display-tag-table table-bordered table-responsive table"})
    table_head = table.find("thead")
```

##### We will get the text data from table head data to further use it for our dataframe 

```python
title = table_head.text
```

```python
title.replace('\n',',')
title=title.strip()
title = title.replace('\n',',')
```

```python
title
```

##### As the type of title is string, we will convert into list for using it directly into our Dataframe later

```python
def ConvertStringToList(string):
    li = list(string.split(","))
    return li
  

table_title_list = ConvertStringToList(title)
```

```python
table_title_list
```

```python
l = []
```

### Using bs4 to navigate the URL and scrape table rows and columns

##### The page range given is for 30 pages and we can give any page count as per our wish

```python
for page in range(1,31): 
   
    req = requests.get("http://aishe.gov.in/aishe/institutionAisheCode?universityId=-1&flag=true&districtCode=-1&universitytypeId=-1&stateId=29&instituteTypeId=-1&institutionType=college&d-4030398-p={}".format(page),verify=False)
    soup = bs(req.text, 'html.parser')
    table = soup.find("table", attrs={"class": "display-tag-table table-bordered table-responsive table"})
    table_data = table.find_all("td")
    row = [tr.text for tr in table_data]
    l.append(row)
```

##### We have got the entire table rows but it has also got the tab spaces and new lines along with it

```python
l
```

##### We will flatten the list first and then remove the /t characters from it

```python
flat_list = [item for sublist in l for item in sublist]
```

```python
flat_list
```

##### Removing all /t,/n and /r characters from list by using split method

```python
c = [''.join(item.split()) for item in flat_list]
c
```

##### We have got the list of all the colleges data but we need to create a list of list as each row consists of 6 columns State, District, University, College, AISHE Code, Status

```python
n=6
chunks = [c[i:i + n] for i in range(0, len(c), n)]
```

```python
chunks
```

##### A chunk of list is created with size of 6 representing a single row table data 

```python
chunks[0]
```

### Final Output (Creating a dataframe)

```python
from pandas import DataFrame

df = DataFrame(chunks,columns=table_title_list)
df
```
